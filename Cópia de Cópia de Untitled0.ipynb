```python
#exercício 7

soma_total = 0
valores = []

while True:
    entrada = input("Digite um número (ou ENTER para encerrar): ")

    if entrada == "":
        break

    try:
        numero = float(entrada)
        valores.append(numero)
        soma_total += numero

        print(f"Valores recebidos até agora: {valores}")
        print(f"Soma parcial: {soma_total}")
        print("-" * 30)
    except ValueError:
        print("Entrada inválida! Digite um número ou ENTER para encerrar.")

print(f"A SOMA TOTAL DOS VALORES FOI: {soma_total}")
```
